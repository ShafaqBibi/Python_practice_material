# Snake, Water, Gun Game

```python
import random

print("Welcome to Snake Water Gun Game\n\nS for Snake\nW for Water\nG for Gun\n\nMake a selection S, W, or G: ")

List = ['s', 'w', 'g']

Compscore = 0
userscore = 0
tie = 0
Round = 10

while Round>0:
    compguess = random.choice(List)
    userguess = input("\nChoose: ")
    
    if compguess == userguess:
        print(f"Round Tie\nYou choose: {userguess}\nComputer choose: {compguess} ")
        tie += 1
        
    elif userguess == 's' and compguess == 'w':
        print(f"You choose: {userguess}\nComputer choose: {compguess}\n\nYou Win")
        userscore += 1
        
    elif userguess == 's' and compguess == 'g':
        print(f"You choose: {userguess}\nComputer choose: {compguess}\n\nComputer Win")
        Compscore += 1
        
    elif userguess == 'w' and compguess == 's':
        print(f"You choose: {userguess}\nComputer choose: {compguess}\n\nComputer Win")
        Compscore += 1
        
    elif userguess == 'w' and compguess == 'g':
        print(f"You choose: {userguess}\nComputer choose: {compguess}\n\nYou Win")
        userscore += 1
        
    elif userguess == 'g' and compguess == 'w':
        print(f"You choose: {userguess}\nComputer choose: {compguess}\n\nComputer Win")
        Compscore += 1
        
    elif userguess == 'g' and compguess == 's':
        print(f"You choose: {userguess}\nComputer choose: {compguess}\n\nYou Win")
        userscore += 1
        
    else:
        print(f"Invalid input")
        break
    
    Round -= 1
    print(f"{Round} rounds left")
        

if userscore > Compscore:
    print(f"\nnYou Win the game\n\nYour Score: {userscore}\nComputer's Score: {Compscore}")
elif userscore < Compscore:
    print(f"\n\nComputer Win the game\n\nYour Score: {userscore}\nComputer's Score: {Compscore}")
else:
    print(f"\n\nGame Tie\n\nYour Score: {userscore}\nComputer's Score: {Compscore}")
```
